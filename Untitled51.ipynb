```python
import numpy as np
from sklearn.model_selection import train_test_split
import os
from PIL import Image

# Load your dataset and labels
Data_Dir = "/content/drive/MyDrive/brain_tumor_dataset"
Cimages = os.listdir(Data_Dir+"/yes")
Nimages = os.listdir(Data_Dir+"/no")

data = []
labels = []

# Load images and labels
for img in Cimages:
    img_path = os.path.join(Data_Dir, "yes", img)
    image = Image.open(img_path).convert('RGB') # Ensure all images are in RGB format
    image = image.resize((224, 224))  # Resize images to 224x224
    image = np.array(image) / 255.0  # Normalize pixel values to [0, 1]
    data.append(image)
    labels.append(1)  # 1 for yes (brain tumor)

for img in Nimages:
    img_path = os.path.join(Data_Dir, "no", img)
    image = Image.open(img_path).convert('RGB') # Ensure all images are in RGB format
    image = image.resize((224, 224))  # Resize images to 224x224
    image = np.array(image) / 255.0  # Normalize pixel values to [0, 1]
    data.append(image)
    labels.append(0)  # 0 for no (no brain tumor)

# Convert data and labels to numpy arrays
data = np.array(data, dtype=np.float32)
labels = np.array(labels, dtype=np.int32)

# Check if data and labels are not empty
if len(data) == 0 or len(labels) == 0:
    raise ValueError("Data or labels is empty!")

# Check if data and labels have the same number of samples
if len(data) != len(labels):
    raise ValueError("Data and labels have different numbers of samples!")

# Split data into training and testing sets
(X_train, X_test, Y_train, Y_test) = train_test_split(data, labels,
                                                      test_size=0.20,
                                                      stratify=labels,
                                                      random_state=42)

print("Data split successfully!")
```

```python
from tensorflow.keras.optimizers import Adam
from sklearn.model_selection import train_test_split
from tensorflow.keras.optimizers.schedules import ExponentialDecay
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dense
from tensorflow.keras.models import Model
from tensorflow.keras.applications import EfficientNetB7
from sklearn.model_selection import train_test_split
import os

# Directory where the dataset is located
img_dir = '/content/drive/MyDrive/brain_tumor_dataset'

# Define parameters
BS = 8
INIT_LR = 1e-3
EPOCHS = 10

# Ensure correct dataset directory structure
# Use proper sub-directory structure: img_dir/class_name/image_files
# Example: /content/drive/MyDrive/brain_tumor_dataset/no/image1.jpg
#          /content/drive/MyDrive/brain_tumor_dataset/yes/image2.jpg

# Define data generators with validation split
# Define data generators with validation split
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # Define validation split
)

val_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # Define validation split
)


train_generator = train_datagen.flow_from_directory(
    img_dir,
    target_size=(224, 224),
    batch_size=BS,
    class_mode='binary',
    subset='training',
    shuffle=True
)

train_idx, val_idx = train_test_split(range(len(train_generator.filenames)), test_size=0.2, random_state=42)

val_generator = val_datagen.flow_from_directory(
    img_dir,
    target_size=(224, 224),
    batch_size=BS,
    class_mode='binary',
    subset='validation',  # Specify this is for validation
    shuffle=False
)

val_generator.filenames = [train_generator.filenames[i] for i in val_idx]


print("Number of training samples:", len(train_generator.filenames))
print("Number of validation samples:", len(val_generator.filenames))

# Check if generators found any images
if len(train_generator.filenames) == 0:
    print("Error: No training images found. Check your directory structure.")
if len(val_generator.filenames) == 0:
    print("Error: No validation images found. Check your validation split and directory structure.")

print("Number of training samples:", len(train_generator.filenames))
print("Number of validation samples:", len(val_generator.filenames))


# Define learning rate schedule
lr_schedule = ExponentialDecay(
    initial_learning_rate=INIT_LR,
    decay_steps=len(train_generator) * EPOCHS,  # Corrected for per epoch steps
    decay_rate=0.1  # Adjust decay rate if needed
)

# Create the Adam optimizer with the learning rate schedule
opt = Adam(learning_rate=lr_schedule)

# Load and modify EfficientNetB7
base_model = EfficientNetB7(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
base_model = EfficientNetB7(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
x = base_model.output
x = tf.keras.layers.GlobalAveragePooling2D()(x)  # Add GlobalAveragePooling2D layer
x = Dense(1, activation="sigmoid")(x)


# Compile the model
model = Model(inputs=base_model.input, outputs=x)
model.compile(loss="binary_crossentropy", optimizer=opt, metrics=["accuracy"])

# Print model summary
model.summary()




```
